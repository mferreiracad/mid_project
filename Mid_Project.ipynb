```python
y = ddata['R_Mortality_2019']
X = ddata.drop(['R_Mortality_2019'],axis=1)
dcontinuous = X[['AVG_Income_2019','Avg_car_age_2019','GDP_PPS_2019','Road_length_2019','n_accidents_by_100k','Educ_H_2019','Alcohol_Consumption_2019']]
dcontinuous
```

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(dcontinuous)
normalized = transformer.transform(dcontinuous)
normalized = pd.DataFrame(normalized,columns=dcontinuous.columns)
normalized
```

```python
# Encode the categorical columns
#first splitting out the columns that need One hot and the ones that need ordinal
dcategoricals = X[['BAC_limit','Happy_Index_2019','Age_to_drive','Urban_speed_limits_2019','Motorway_speed_limits_2019']]
one_hot_enc = dcategoricals[['BAC_limit','Age_to_drive','Urban_speed_limits_2019','Motorway_speed_limits_2019']]
ordinal_enc = dcategoricals['Happy_Index_2019']
```

```python
#One hot encoding first
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder().fit(one_hot_enc)
encoded_onehot = encoder.transform(one_hot_enc).toarray()

encoded_columns = list()
dropped_columns = list()
for columns in encoder.categories_:
    condition = True
    for column in columns.tolist():
        if condition:
            dropped_columns.append(column)
            encoded_columns.append(column)
            condition = False
        else:
            encoded_columns.append(column)

onehot_encoded = pd.DataFrame(data = encoded_onehot, columns=encoded_columns)
onehot_encoded = onehot_encoded.drop(columns = dropped_columns)

onehot_encoded
```

```python
#The ordinal, the position in the hapiness index, don't need further encoding
```

```python
X = pd.concat([ordinal_enc,onehot_encoded,normalized],axis=1)
X.head()
```

```python
#seperate the data with a train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
from sklearn import linear_model
lm = linear_model.LinearRegression()
model= lm.fit(X_train,y_train)
```

```python
score = lm.score(X_test, y_test)
score
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
def model_training(models,X_train,y_train):
    models_train = []
    for model in models:
        model.fit(X_train,y_train)
        models_train.append(model)
    return models_train
        
        
trained = model_training([MLPRegressor(max_iter=1000),linear_model.LinearRegression(),KNeighborsRegressor()],X_train,y_train)
```

```python
#scores
score_mlp = trained[0].score(X_test, y_test)
score_linear = trained[1].score(X_test, y_test)
score_KNN = trained[2].score(X_test, y_test)

print('MLPRegressor Score:',score_mlp)
print('LinearRegressor Score:',score_linear)
print('KNN Score:',score_KNN)
```

```python
from scipy import stats
stats_value, pvalue = stats.ttest_1samp(age,6)
print('Stat_Score =  ',stats_value)
print('Pvalue =  ',pvalue/2)
```

```python
import scipy.stats

confidence_level = 0.95
degrees_freedom = len(age) - 1  # or, for large values of sample size, just use that
sample_mean = np.mean(age)
# note that we use the standard error of the sample 
# as an estimate of the standard error of the population (which is used in the theoretical formula)
sample_standard_error = scipy.stats.sem(age) # sem = standard error of the mean = std(mean)/sqrt(samplesize)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```
